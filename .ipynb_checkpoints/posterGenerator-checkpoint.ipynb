## 2.포스터 프레임 추출

##### OpenCV와 YOLOv5를 사용한 프레임 추출

```python
import cv2
import random
import matplotlib.pyplot as plt

# Haar Cascade 모델을 사용하여 얼굴 감지
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def select_best_face_frame(video_path, output_path=None):
    # 영상 파일 열기
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # 무작위로 10개의 프레임 선택
    random_frames = random.sample(range(total_frames), 20)
    
    best_frame = None
    highest_score = 0  # 얼굴 인식 점수 기록
    best_frame_no = None  # 최적 프레임 번호 기록
    
    for frame_no in random_frames:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_no)
        ret, frame = cap.read()
        if not ret:
            continue

        # 얼굴 인식 수행
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # 얼굴 크기를 기준으로 점수 부여 (얼굴 영역의 총 크기)
        score = sum((w * h) for (x, y, w, h) in faces)

        # 최고 점수인 경우 해당 프레임을 선택
        if score > highest_score:
            highest_score = score
            best_frame = frame
            best_frame_no = frame_no

    # 선택된 프레임의 주변 10개 프레임 탐색
    if best_frame_no is not None:
        best_sharp_frame = best_frame
        highest_sharpness = 0  # 선명도 기록
        
        for i in range(best_frame_no - 10, best_frame_no + 11):
            if i < 0 or i >= total_frames:
                continue  # 프레임 범위를 벗어나면 무시

            cap.set(cv2.CAP_PROP_POS_FRAMES, i)
            ret, frame = cap.read()
            if not ret:
                continue

            # 선명도 측정 (Laplacian 변환을 이용한 에지 측정)
            gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            laplacian_var = cv2.Laplacian(gray_frame, cv2.CV_64F).var()
            
            # 선명도가 더 높은 경우 프레임 갱신
            if laplacian_var > highest_sharpness:
                highest_sharpness = laplacian_var
                best_sharp_frame = frame

        best_frame = best_sharp_frame

    cap.release()
    
    # 프레임 저장 (output_path가 지정된 경우)
    if best_frame is not None and output_path:
        cv2.imwrite(output_path, best_frame)
        print(f"프레임이 {output_path}에 저장되었습니다.")
    
    return best_frame

# 예제 실행
output_file = 'ProcessedData/selected_frame.jpg'  # 저장할 파일 경로
frame = select_best_face_frame('MovieData/Blooming.mp4', output_path=output_file)

if frame is not None:
    # BGR 이미지를 RGB로 변환
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame_rgb)
    plt.axis('off')  # 축 숨기기
    plt.show()
else:
    print("적절한 프레임을 찾을 수 없습니다.")
```

## 3.포스터 오토 레이아웃

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from ultralytics import YOLO
from PIL import Image, ImageDraw, ImageFont, ImageFilter
import joblib
```

```python
# 하위 폴더에서 모델 불러오기
regressor = joblib.load('Poster/poster_model.pkl')  # 상위 폴더에서 모델을 불러옴
```

```python
# YOLO 모델 로드
yolo_model = YOLO("yolov8n.pt")  # YOLOv8n 모델 다운로드

# 인물 위치 탐지 함수
def detect_person(image_path):
    results = yolo_model.predict(source=image_path)  # YOLO 예측
    detections = results[0].boxes  # 첫 번째 이미지 결과의 Bounding Boxes

    person_boxes = []
    for box in detections:
        coords = box.xyxy[0].tolist()  # Bounding Box 좌표: [x1, y1, x2, y2]
        class_id = int(box.cls[0])  # 클래스 ID
        if class_id == 0:  # 클래스 0: "person"
            person_boxes.append(coords)
    return person_boxes

# 텍스트 위치 예측 함수
def predict_text_positions(person_boxes):
    text_positions = []
    for box in person_boxes:
        x1, y1, x2, y2 = map(int, box)
        prediction = regressor.predict([[x1, y1, x2, y2]])  # Random Forest 예측
        text_positions.append(prediction[0])  # 예측된 텍스트 좌표
    return text_positions

# 텍스트 위치 예측 함수 (멀고 큰 텍스트 박스 선택)
def predict_best_text_position(person_boxes):
    best_text_position = None
    best_score = -float("inf")  # 초기값: 매우 작은 값

    for box in person_boxes:
        x1_person, y1_person, x2_person, y2_person = map(int, box)
        person_center_x = (x1_person + x2_person) / 2
        person_center_y = (y1_person + y2_person) / 2

        for text_box in text_positions:
            tx1, ty1, tx2, ty2 = map(int, text_box)

            # 텍스트 중심 좌표 계산
            text_center_x = (tx1 + tx2) / 2
            text_center_y = (ty1 + ty2) / 2

            # 인물 중심과 텍스트 중심 거리 계산 (유클리드 거리)
            distance = ((text_center_x - person_center_x) ** 2 + (text_center_y - person_center_y) ** 2) ** 0.5

            # 텍스트 박스 크기 계산
            text_width = tx2 - tx1
            text_height = ty2 - ty1
            text_area = text_width * text_height

            # 점수 계산 (가중치 조정 가능)
            score = distance + 0.5 * text_area  # 거리와 크기를 조합한 점수

            # 가장 높은 점수의 텍스트 박스 선택
            if score > best_score:
                best_score = score
                best_text_position = [tx1, ty1, tx2, ty2]

    return best_text_position



def insert_text(image, text, text_box, font_path="/System/Library/Fonts/Supplemental/Helvetica.ttc", font_size=30):
    """
    선택된 텍스트 박스에 텍스트를 삽입하는 함수
    Args:
        image: PIL Image 객체
        text: 삽입할 텍스트
        text_box: 텍스트가 삽입될 박스 좌표 [x1, y1, x2, y2]
        font_path: 폰트 경로
        font_size: 폰트 크기
    Returns:
        텍스트가 삽입된 이미지
    """
    draw = ImageDraw.Draw(image)
    tx1, ty1, tx2, ty2 = map(int, text_box)

    # 텍스트 영역 크기
    box_width = tx2 - tx1
    box_height = ty2 - ty1

    # 폰트 로드
    try:
        font = ImageFont.truetype(font_path, font_size)
    except Exception as e:
        print(f"Font loading error, using default font: {e}")
        font = ImageFont.load_default()

    # 텍스트 크기 계산
    bbox = font.getbbox(text)  # 텍스트 경계 박스 계산
    text_width, text_height = bbox[2] - bbox[0], bbox[3] - bbox[1]

    # 텍스트 크기가 박스 크기보다 크면 폰트 크기 줄이기
    while text_width > box_width or text_height > box_height:
        font_size -= 2  # 폰트 크기 줄이기
        font = ImageFont.truetype(font_path, font_size)
        bbox = font.getbbox(text)
        text_width, text_height = bbox[2] - bbox[0], bbox[3] - bbox[1]

    # 텍스트를 텍스트 박스의 중앙에 배치
    text_x = tx1 + (box_width - text_width) // 2
    text_y = ty1 + (box_height - text_height) // 2

    # 텍스트 삽입
    draw.text((text_x, text_y), text, fill="white", font=font)
    return image

```

```python
# 테스트 이미지
new_image_path = "ProcessedData/selected_frame.jpg"  # 테스트 이미지 경로
title_text = excel_data.loc[excel_data['항목'] == '작품명 (영문)'].values[0][1]
person_boxes = detect_person(new_image_path)  # 인물 탐지
text_positions = predict_text_positions(person_boxes)  # 텍스트 예측
best_text_box = predict_best_text_position(person_boxes)  # 가장 적합한 텍스트 박스 선택


# 이미지 시각화
image = Image.open(new_image_path)
plt.imshow(image)
ax = plt.gca()


# 인물 박스 및 텍스트 박스 그리기
for box in person_boxes:
    x1, y1, x2, y2 = map(int, box)
    rect = patches.Rectangle((x1, y1), x2 - x1, y2 - y1, edgecolor="green", facecolor="none", linewidth=2)
    ax.add_patch(rect)

for text_box in text_positions:
    tx1, ty1, tx2, ty2 = map(int, text_box)
    rect = patches.Rectangle((tx1, ty1), tx2 - tx1, ty2 - ty1, edgecolor="blue", facecolor="none", linewidth=2)
    ax.add_patch(rect)

# 베스트 텍스트 박스와 텍스트 삽입
if best_text_box:
    tx1, ty1, tx2, ty2 = map(int, best_text_box)
    rect = patches.Rectangle((tx1, ty1), tx2 - tx1, ty2 - ty1, edgecolor="yellow", facecolor="none", linewidth=2)
    ax.add_patch(rect)

    # 텍스트 삽입
    title = 'Blooming'
    font_path = "/Library/Fonts/Noteworthy.ttc"
    font_size = 100
    image_with_text = insert_text(image.copy(), title, best_text_box, font_path, font_size)

    # 텍스트 삽입된 이미지 표시
    plt.imshow(image_with_text)
    plt.axis("off")
    plt.title("Image with Text")
    plt.show()
    image_with_text.save("ProcessedData/poster_with_title.jpg")
else:
    print("No suitable text box found.")
    plt.imshow(image)
    plt.axis("off")
    plt.title("Original Image")
    plt.show()
```
