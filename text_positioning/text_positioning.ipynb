# 텍스트 배치
##### 선택된 프레임에 학습된 모델을 이용하여 텍스트를 배치한다.
##### 선택된 프레임을 먼저 라벨링 후 학습해야함. 
##### 모델을 사용할 때 640*360이미지를 인풋으로 넣어야 한다. 따라서 모델 사용 전 리사이징하고 모델 사용 후 복원해야한다.(추후 더 많은 포스터를 학습시켜 다양한 인풋을 처리하도록 학습 필요)

## 1. 선택된 포스터 인물 라벨링
##### poster_frame_selector/selected_frames폴더의 이미지를 Yolo v8로 인물을 라벨링하여 df_labeled_data변수에 저장한다

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
from ultralytics import YOLO
from PIL import Image
import logging
import joblib
```

```python
# 로깅 설정: YOLO의 출력 메시지를 ERROR 레벨로 설정하여 정보성 메시지 숨기기
logging.getLogger("ultralytics").setLevel(logging.ERROR)

# YOLOv8 모델 로드
yolo_model = YOLO("yolov8n.pt", verbose=False)  # YOLOv8 모델
yolo_model.model.conf = 0.6  # Confidence threshold 설정 (필요시)
```

```python
image_folder = "../poster_frame_selector/selected_frames"
label_data = []  # 결과 저장 리스트
image_files = [f for f in os.listdir(image_folder) if os.path.isfile(os.path.join(image_folder, f))]

for idx, image_file in enumerate(image_files):
    image_path = os.path.join(image_folder, image_file)

    # 이미지 로드
    image = cv2.imread(image_path)
    if image is None:
        print(f"\tCannot read {idx}: {image_file}")
        continue

    # YOLOv8로 인물 위치 검출
    results = yolo_model(image_path)
    person_boxes = []
    for box in results[0].boxes.data.tolist():  # YOLOv8 바운딩 박스 가져오기
        x1, y1, x2, y2, conf, cls = box[:6]
        if int(cls) == 0:  # 클래스 ID 0: "person"
            person_boxes.append((x1, y1, x2, y2, conf))

    # 신뢰도가 가장 높은 바운딩 박스 선택
    if person_boxes:  # person_boxes가 비어있지 않은 경우에만 실행
        best_box = max(person_boxes, key=lambda x: x[4])  # 신뢰도(conf) 기준으로 가장 높은 박스 선택
        x1, y1, x2, y2, conf = best_box

        # 라벨 데이터 저장
        label_data.append({
            "Poster_File": image_file,
            "Object_Type": "Character",
            "X1_char": int(x1),
            "Y1_char": int(y1),
            "X2_char": int(x2),
            "Y2_char": int(y2),
            "Confidence": round(conf, 2)
        })

# 라벨링 결과 확인
df_labeled_data = pd.DataFrame(label_data)
df_labeled_data
```

## 2. 텍스트 위치 설정
##### 학습한 모델를 활용하여 인물의 바운딩박스로 텍스트 위치를 설정한다.
##### train_poster_model/text_locator.pkl파일을 활용하여 리사이징한 df_labeled_data를 입력하고 text_position을 출력으로 받는다. 그 후, 사이즈를 다시 복원하여 df_selected변수에 입출력 값을 저장한다.

```python
# 입력 데이터 준비
X = df_labeled_data[['X1_char', 'Y1_char', 'X2_char', 'Y2_char']].values  # Character 좌표

# 좌표 리사이징: 모델에 학습시킬때는 640*360으로 변환시켜야함.
height, width, channels = image.shape # 이미지 크기 가져오기
X_resizer = 640 / width
Y_resizer = 360 / height

# 리사이즈된 좌표 계산
X[:, [0, 2]] = X[:, [0, 2]] * X_resizer  # x 좌표(X1_char, X2_char)
X[:, [1, 3]] = X[:, [1, 3]] * Y_resizer  # y 좌표(Y1_char, Y2_char)

# 모델 불러오기
model = joblib.load('../train_poster_model/text_locator.pkl')

# 텍스트 좌표 예측
y_pred = model.predict(X)

# 결과 확인
text_position = pd.DataFrame(y_pred.round(0).astype(int))
text_position
```

```python
# 결과 데이터에 컬럼 이름 지정
text_position.columns = ['X1_text', 'Y1_text', 'X2_text', 'Y2_text']

# 리사이즈된 좌표를 원본 좌표로 복원
text_position['X1_text'] = (text_position['X1_text'] / X_resizer).astype(int)
text_position['Y1_text'] = (text_position['Y1_text'] / Y_resizer).astype(int)
text_position['X2_text'] = (text_position['X2_text'] / X_resizer).astype(int)
text_position['Y2_text'] = (text_position['Y2_text'] / Y_resizer).astype(int)

# 입력 데이터와 결과 데이터 병합
df_combined = pd.concat([df_labeled_data, text_position], axis=1)

# 필요한 열만 선택하여 저장 후 결과 출력
df_selected = df_combined[['Poster_File', 'X1_char', 'Y1_char', 'X2_char', 'Y2_char', 'X1_text', 'Y1_text', 'X2_text', 'Y2_text']]
df_selected
```

```python
# 색상 및 두께 설정
character_color = (255, 0, 0)  # 파란색 (BGR)
text_color = (0, 255, 0)       # 초록색 (BGR)
line_thickness = 5
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 2
font_thickness = 2

# 라벨링 표시 함수
def draw_labels(image, row):
    # Character 좌표
    x1_char, y1_char, x2_char, y2_char = row[['X1_char', 'Y1_char', 'X2_char', 'Y2_char']]
    # Text 좌표
    x1_text, y1_text, x2_text, y2_text = row[['X1_text', 'Y1_text', 'X2_text', 'Y2_text']]
    
    # 인물(Character) 사각형과 라벨
    cv2.rectangle(image, (x1_char, y1_char), (x2_char, y2_char), character_color, line_thickness)
    cv2.putText(image, 'Character', (x1_char, y1_char - 10), font, font_scale, character_color, font_thickness)
    
    # 텍스트(Text) 사각형과 라벨
    cv2.rectangle(image, (x1_text, y1_text), (x2_text, y2_text), text_color, line_thickness)
    cv2.putText(image, 'Text', (x1_text, y1_text - 10), font, font_scale, text_color, font_thickness)
    
    return image

# 이미지 파일 처리
processed_images = {}  # 이미지 파일별 처리 결과 저장
for idx, row in df_selected.iterrows():
    poster_file = row['Poster_File']
    image_path = os.path.join(image_folder, poster_file)
    
    # 이미지 읽기
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        continue
    
    # 이미지가 이미 처리되지 않은 경우에만 처리
    if poster_file not in processed_images:
        # 새로운 이미지 파일 읽기
        image = cv2.imread(image_path)
        
        # 라벨링 추가
        labeled_image = draw_labels(image, row)
        processed_images[poster_file] = labeled_image  # 처리 결과 저장

# 시각화
fig, axes = plt.subplots(len(processed_images), 1, figsize=(10, 20))

for i, (poster_file, img) in enumerate(processed_images.items()):
    # OpenCV 이미지를 RGB로 변환
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img_rgb)
    axes[i].axis('off')
    axes[i].set_title(f"{poster_file}")

plt.tight_layout()
plt.show()
```

# 3. 텍스트 배치 및 저장
##### movie_data의 작품명을 가져와 이미지와 합성시킨 후 frames_with_text폴더에 저장한다.

```python
# 설정
movie_data = pd.read_csv('movie_data.csv') # 영화 데이터 불러오기
output_folder = "./frames_with_text" # 이미지 저장 폴더 경로
movie_title = movie_data[movie_data['항목'] == '작품명 (영문)']['내용'].iloc[0] # 영화 타이틀 가져오기

# 텍스트 색상 및 두께 설정
font = cv2.FONT_HERSHEY_SIMPLEX
font_thickness = 50
text_color = (255, 255, 255)

# 텍스트 위치 및 크기 계산
def calculate_text_position(row):
    # 텍스트 중앙 좌표 계산
    x_position = int((row['X2_text'] + row['X1_text']) / 2)
    y_position = int((row['Y2_text'] + row['Y1_text']) / 2)
    
    # 폰트 크기 계산
    font_size = (row['Y2_text'] - row['Y1_text']) * 0.02  # float로 사용
    return x_position, y_position, font_size


# 텍스트 추가 함수
def add_title(image, title, x, y, font_size):
    text_size = cv2.getTextSize(title, font, font_size, font_thickness)[0]
    text_x = x - text_size[0] // 2
    text_y = y + text_size[1] // 2
    cv2.putText(image, title, (text_x, text_y), font, font_size, text_color, font_thickness)
    return image

# 이미지 처리 및 텍스트 추가
processed_images = {}
for idx, row in df_selected.iterrows():
    poster_file = row['Poster_File']
    image_path = os.path.join(image_folder, poster_file)
    
    # 이미지 읽기
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        continue

    image = cv2.imread(image_path)
    if image is None:
        print(f"Could not load image: {poster_file}")
        continue
    
    # 텍스트 위치 및 크기 계산
    x_pos, y_pos, font_size = calculate_text_position(row)

    # 영화 제목 추가
    labeled_image = add_title(image, movie_title, x_pos, y_pos, font_size)
    processed_images[poster_file] = labeled_image

    # 저장 경로 설정
    output_path = os.path.join(output_folder, poster_file)

    # 이미지 저장
    cv2.imwrite(output_path, labeled_image)
    print(f"Saved: {output_path}")

# 시각화
fig, axes = plt.subplots(len(processed_images), 1, figsize=(10, 20))

for i, (poster_file, img) in enumerate(processed_images.items()):
    # OpenCV 이미지를 RGB로 변환
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img_rgb)
    axes[i].axis('off')
    axes[i].set_title(f"{poster_file}")

plt.tight_layout()
plt.show()
```
